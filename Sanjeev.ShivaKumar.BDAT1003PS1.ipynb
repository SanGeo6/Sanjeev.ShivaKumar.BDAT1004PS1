{

  {
    string str = "Supercalifragilisticexpialidocious";
    int count = 0;
    for(int i=0;i<str.Length;i++)
    {   
       const char V = '\0';
       if (str[i] != V)
           count++;
    }
    Console.WriteLine("Total number of characters in the given string = " + count); 
      
  }
}

using System;
using System.Text;

public class myTriangle
{
    static float myArea(float a, float b, float c)
    {
        float s = (a + b + c) / 2;
        return (float)Math.Sqrt(s * (s - a) * (s - b) * (s - c));
    }
    
    public static void Main(String[] args)
    {
     Console.Write("Enter the value of sides of the Triangle : ");
     float a = float.Parse(Console.ReadLine());
     float b = float.Parse(Console.ReadLine());
     float c = float.Parse(Console.ReadLine());
     Console.Write("Area of the Triangle using Heron's formula is " + myArea(a, b, c)); 
}
}
